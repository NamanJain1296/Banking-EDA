# Box Plot

```python
for x in wd.columns:
    plt.figure(figsize = (8,5))
    wd.boxplot(column = [x])
    plt.title(f'Box Plot for {x}')
    plt.ylabel('Values')
```

# Histogram

```python
for x in wd.columns:
    wd.hist(column = [x], color = 'purple')
    plt.title(f'Box Plot for {x}')
```

- There are no missing values inside the dataset

# Scatter Plot

```python
col = ['Type', 'Alcohol', 'Malic', 'Ash', 'Alcalinity', 'Magnesium',
       'Phenols', 'Flavanoids', 'Nonflavanoids', 'Proanthocyanins',
       'Color', 'Hue', 'Dilution', 'Proline']

for i in range(len(col) - 1):
    x = col[i]
    y = col[i+1]
    plt.scatter(x = wd[x], y = wd[y], color = 'green')
    plt.title(f'Scatter Plot for {x} vs {y}')
    plt.xlabel(x)
    plt.ylabel(y)
    plt.show()
```

# Pie Chart

```python
types = wd['Type'].value_counts()
types.plot.pie(autopct = '%1.1f%%', startangle = 90, figsize = (8,8), title = 'Types of Each Number')
```

# Bar Graph

```python
types = wd['Type'].value_counts()
types.plot.bar(color = 'red')
```
